Class Distrubition Plot

```python
sns.set(style="whitegrid")

# Prepare the data
df_plot = df.copy()
df_plot['Class'] = df_plot['Class'].map({0: 'No Tumor', 1: 'Tumor'})  # Optional mapping for readability

plt.figure(figsize=(6, 4))

# Change palette to blue and gray
bars = sns.countplot(data=df_plot, x='Class', palette=["#429efb", "#bcc2c9"], hue='Class', legend=False)

# Add annotations
for bar in bars.patches:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f'{int(height)}',
        ha='center',
        va='bottom',
        fontsize=12,
        color='black'
    )

plt.xlabel('')
plt.ylabel('Number of Samples', fontsize=12)
plt.title('Class Distribution: Tumor vs No Tumor', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.savefig('class_distribution.png', dpi=300)  # Save figure automatically
plt.show()
```

Display Some Images Tumor vs No Tumor

```python
# Charger le dataset
image_dir = "Brain_Tumor_Imges"

# Sélectionner 6 images aléatoires par classe
tumor_images = df[df['Class'] == 1]['Image'].sample(6, random_state=42).tolist()
no_tumor_images = df[df['Class'] == 0]['Image'].sample(6, random_state=42).tolist()

# Fonction pour retrouver le chemin réel de l’image (avec extension auto)
def find_image_path(image_name, folder):
    pattern = os.path.join(folder, image_name + '.*')
    matches = glob.glob(pattern)
    return matches[0] if matches else None

# Fonction pour afficher les images
def display_images(image_list, label, start_pos):
    for i, img_name in enumerate(image_list):
        img_path = find_image_path(img_name, image_dir)
        if img_path:
            image = mpimg.imread(img_path)
            plt.subplot(2, 6, start_pos + i)
            plt.imshow(image, cmap='gray')
            plt.title(label, fontsize=10)
            plt.axis('off')
        else:
            print(f"Image not found for: {img_name}")

# Affichage des images
plt.figure(figsize=(18, 6))
display_images(tumor_images, "Tumor", 1)
display_images(no_tumor_images, "No Tumor", 7)
plt.suptitle("Tumor vs No Tumor Images", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Leave space for the suptitle

# === Save as one image ===
plt.savefig("tumor_vs_no_tumor_samples.png", dpi=300)
plt.show()
```

Features Scaling

```python
# Save Image, Class and features separately
image_col =df["Image"] #Stores the images names for reference traceability
X = df.drop(columns=["Image", "Class"]) #Only the features (without class & image)
Y = df["Class"] #The class
```

Features Correlation

```python
corr = df[X.columns].corr()
plt.figure(figsize=(6, 4))
sns.heatmap(corr, cmap="coolwarm", annot=False)
plt.title("Feature Correlation Matrix")
plt.show()
```

Standardize The Features

```python
scaler = StandardScaler()                     # Initialize scaler
X_scaled = scaler.fit_transform(X)            # Scale features
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)  # Convert Back to DataFrame
```

Save The Preprocessed Datasets

```python
# Combine the features (scaled) and the target labels (Class)
df_svm_data = pd.DataFrame(X_scaled, columns=X.columns)  # Scaled features
df_svm_data['Class'] = Y  # Add the target 'Class' column
# Save the preprocessed data (features and labels) to a CSV file
df_svm_data.to_csv("Final_svm_preprocessed_data.csv", index=False)
```

Split The Datasets Into Training/test

```python
#20% test and 80% Training
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled_df, Y,
    test_size=0.2, # 20%  Testing
    random_state=42,
    stratify=Y # Maintain the same proportion of classes (Tumor/No Tumor) in train and test
)
```

Class Distribution of the Training/Test Datasets

```python
# Create a DataFrame for training set distribution
train_dist = pd.DataFrame({
    'Class': y_train.value_counts().index,
    'Count': y_train.value_counts().values,
    'Percentage %': (y_train.value_counts(normalize=True).values * 100).round(2)
})

# Create a DataFrame for testing set distribution
test_dist = pd.DataFrame({
    'Class': y_test.value_counts().index,
    'Count': y_test.value_counts().values,
    'Percentage %': (y_test.value_counts(normalize=True).values * 100).round(2)
})
# Display the training set distribution
print("\nTraining Set Distribution:")
print(train_dist.to_string(index=False))

# Display the testing set distribution
print("\nTesting Set Distribution:")
print(test_dist.to_string(index=False))
```

Plot Distribution of Training\Test Datasets

```python
# Combine data for plotting
train_labels = y_train.copy()
test_labels = y_test.copy()

train_df = pd.DataFrame({'Set': 'Training', 'Class': train_labels})
test_df = pd.DataFrame({'Set': 'Testing', 'Class': test_labels})

plot_df = pd.concat([train_df, test_df])

# Optional: map class labels for better readability
plot_df['Class'] = plot_df['Class'].map({0: 'No Tumor', 1: 'Tumor'})

# Plot
plt.figure(figsize=(7, 5))
sns.set(style="whitegrid")
ax = sns.countplot(data=plot_df, x='Set', hue='Class', palette=["#bcc2c9", "#429efb"])
# Add annotations
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=10)

plt.title('Class Distribution in Training and Testing Sets', fontsize=14)
plt.ylabel('Number of Samples')
plt.xlabel('')
plt.legend(title='Class')
plt.tight_layout()
plt.savefig("class_distribution_train_test.png", dpi=300)
plt.show()
```

Save The Training/Test Datas

```python
# Combine training features and labels
train_df = pd.DataFrame(X_train, columns=X.columns)
train_df['Class'] = y_train.values
# Combine testing features and labels
test_df = pd.DataFrame(X_test, columns=X.columns)
test_df['Class'] = y_test.values
# Save to CSV
train_df.to_csv("Final_train_data.csv", index=False)
test_df.to_csv("Final_test_data.csv", index=False)
```
